```python
import pandas as pd
```

```python
market1customers = pd.read_json(r"Market 1 Customers.json")
market1deliveries = pd.read_csv(r"Market 1 Deliveries.csv")
market1orders = pd.read_csv(r"Market 1 Orders.csv")
market2customers = pd.read_json(r"Market 2 Customers.json")
market2deliveries = pd.read_csv(r"Market 2 Deliveries.csv")
market2orders = pd.read_csv(r"Market 2 Orders.csv")
```

```python
mergedcustomers = pd.merge(market1customers,market2customers,on='Customer ID',how='outer')
mergedorders = pd.merge(market1orders,market2orders,on='Order ID',how='outer')
mergeddeliveries = pd.merge(market1deliveries,market2deliveries,on='Task_ID',how='outer')
```

```python
df = mergedorders
df.to_csv("df.csv",index=False)
```

```python
mergedcustomers.to_csv("mergedcustomers.csv",index=False)
mergedorders.to_csv("mergedorders.csv",index=False)
mergeddeliveries.to_csv("mergeddeliveries.csv",index=False)

```
